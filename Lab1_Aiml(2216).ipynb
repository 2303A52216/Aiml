```python
# PART 2
# Visualization of Data using matplotlib, pyplots Packages.
# 1. Generate a Data set for Health Care using the following:

people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
plt.title("Relation between weight and height")
plt.xlabel("weight in kilograms")
plt.ylabel("height in cm")
plt.scatter(weight,height)
plt.show()
```

```python
#Bar Graph
plt.xlabel("people",fontsize=15)
plt.ylabel("Weight in kg",fontsize=15)
plt.bar(people,weight)
plt.show()
```

```python
#Provide Graph Title, labels for X, Y axis with proper justi cation and explanation of
#the graph
plt.figure(figsize=(15,5))
plt.title("peoples height and weight",fontsize=15,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.5,color='white',alpha=1,edgecolor='k')
plt.xlabel("people",fontsize=15)
plt.ylabel("Weight in kg",fontsize=15)
```

```python
# Histogram
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='orange',edgecolor='k',alpha=0.75,bins=3)
plt.xlabel("Weight in kg",fontsize=15)
plt.title("Histogram of patient weight",fontsize=15)
plt.show()
```
